# My role is to cleanse the customer list by excluding those who have indicated a 'do not call' preference, preparing a clean dataset for the client's use in their calling initiatives.

# Let's begin..

# Remove Duplicates
I always begin with this step..

```python
df = df.drop_duplicates()
df
```

# Dropping "Not_Useful_column"

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

# Using "strip" function to clean "last_name" column because it's having some extra characters,
### and if you observe carefully we have to replace values which are on both left and right side. (.,/,_). That's why I am using 'Strip' function because it can address element on both the left and right side simultaneously.

```python
df["Last_Name"]=df["Last_Name"].str.strip("123._/")
df
```

# Cleaning 'Phone_Numbers' column as most of the numbers are in different format.


### As a first step I'll be getting rid of all extra character between them(/, -, |) and convert them in completely numeric format and then I'll format them like this. "000-000-0000". 

```python
df['Phone_Number']= df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df
```

## I got this following error because, if you observe carefully in 'Phone_Numbers' Column there's no proper value, we have number, string (Na), blanks,etc. So what I'll do is convert the data type of this column to 'string' and then apply this function.

```python
df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
```

### Converting to string

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:])
df
```

```python
df['Phone_Number']= df['Phone_Number'].str.replace('nan--','')
df['Phone_Number']= df['Phone_Number'].str.replace('Na--','')

df
```

## Phone_Number column sucessfully formatted!

# Cleaning 'Address' Column
### as it is jumbled with zip code, street address and also state name.
### So I am going to split it into 3 columns i.e 'Street_address', 'State', 'Zip_code'.

```python
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',',2,expand = True)
```

```python
df
```

# Formatting 'Paying customer' and 'Do not contact'' column

```python
df['Paying Customer']=df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer']=df['Paying Customer'].str.replace('No','N')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact']=df['Do_Not_Contact'].str.replace('No','N')
df
```

# Treating nulls

```python
df = df.fillna('')
df
```

# Eliminating customer's details who said 'Yes' in 'Do_Not_Contact' Column 

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

# Eliminating customer's details who have not provided their Phone numbers.

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

# Reseting Index

```python
df = df.reset_index(drop=True)
df
```

# Dropping old 'Address' column and 'zip code'(bcoz zip code is blank)

```python
df = df.drop(columns = ['Address', 'Zip_Code'])
df
```

# The data has been cleaned  and it now features a list of customers open to future calls.
